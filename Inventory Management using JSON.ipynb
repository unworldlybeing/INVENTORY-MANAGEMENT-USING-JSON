```python
import time
import json

# Record.json file read kar rahe hain
fd=open("Record.json",'r')
js=fd.read()
fd.close()
sales=''

# JSON ko Python dictionary mein convert kar rahe hain
record=json.loads(js)

# Product menu display karna
print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['Name'], record[key]['Price'], record[key]['Name'])
print("---------------------------------------------")
print('')

# User se details input lena
ui_name=str(input("Enter your name: "))
ui_mail=str(input("Enter your Mail-ID: "))
ui_phone=str(input("Enter your phone number: "))
ui_pr = input("Enter product ID : ")
ui_qn = int(input("Enter Quantity   : "))

print("---------------------------------------------")
print('')

# Check kar rahe hain ki required quantity available hai ya nahi
if record[ui_pr]['Qn'] >= ui_qn:
    if ui_pr in record:
        # Quantity update kar rahe hain
        record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn

        # Billing info print karna
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["Price"])
        print("Quantity  : ", ui_qn)
        print("---------------------------------------------")
        print("Billing   : ", ui_qn * record[ui_pr]["Price"], "Rs")
        print("---------------------------------------------")

        # Sales ka data prepare karna
        sales=ui_name+","+ui_mail+","+ui_phone+","+ui_pr+","+record[ui_pr]['Name']+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]['Price'])+","+str(record[ui_pr]['Qn']*record[ui_pr]['Price'])+","+time.ctime()+"\n"

else:
    # Agar quantity available nahi hai
    print("SORRY WE ARE NOT HAVING ENOUGH QUANTITY")
    print("WE ARE ONLY HAVING", record[ui_pr]['Qn'],"QUANTITY")
    print("---------------------------------------------")

    # Customer se confirm lena agar woh available quantity lena chahta hai
    ch=str(input("Press Y/y to purchase"))
    if(ch=='Y' or ch=='y'):
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["Price"])
        print("Quantity  : ", record[ui_pr]['Qn'])
        print("---------------------------------------------")
        print("Billing   : ", ui_qn * record[ui_pr]["Price"], "Rs")
        print("---------------------------------------------")

        # Sales ka data prepare karna (available quantity ke liye)
        sales=ui_name+","+ui_mail+","+ui_phone+","+ui_pr+","+record[ui_pr]['Name']+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]['Price'])+","+str(record[ui_pr]['Qn']*record[ui_pr]['Price'])+","+time.ctime()+"\n"
        
        # Product ki quantity 0 set kar rahe hain
        record[ui_pr]['Qn'] =0
    else:
        print("Thanks!")

# Record.json file update karna (inventory update)
js = json.dumps(record)
fd = open('Record.json','w')
fd.write(js)
fd.close()

# Sales record file mein likhna
fd=open("salesJson.txt",'a')
fd.write(sales)
fd.close()

print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")        
```
